```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import os
from mido import MidiFile, MetaMessage
from mido.midifiles.meta import KeySignatureError
from tqdm import tqdm
import time

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id='',
                                                           client_secret=''))

######################@


 # Function to extract artist and song title from file path
def extract_artist_song(file_path, base_folder):
    relative_path = os.path.relpath(file_path, base_folder)
    parts = relative_path.split(os.sep)
    artist = parts[0]
    song_title = os.path.splitext(parts[1])[0]
    return artist, song_title








##############@

def find_genre_spotify(artist, song_title):
    search_query = f'artist:{artist} track:{song_title}'
    result = sp.search(search_query, type='track', limit=1)

    if result['tracks']['total'] > 0:
        track = result['tracks']['items'][0]
        artist_id = track['artists'][0]['id']
        artist_data = sp.artist(artist_id)

        if artist_data.get('genres'):
            return ', '.join(artist_data['genres'])

    return None


##################

def add_genre_to_midi1(file_path, genre):
    try:
        midi_file = MidiFile(file_path)
    except IOError as e:
        print(f"Error reading MIDI file {file_path}: {e}")
        return
    except EOFError as e:
        print(f"Unexpected end of MIDI file {file_path}: {e}")
        return
    except KeySignatureError as e:
        print(f"Key signature error in MIDI file {file_path}: {e}")
        return
    except ValueError as e:
        print(f"ValueError in MIDI file {file_path}: {e}")
        return
    except Exception as e:
        print(f"Unexpected error in MIDI file {file_path}: {e}")
        return

    for track in midi_file.tracks:
        track.append(MetaMessage('text', text=f"Genre: {genre}"))

    midi_file.save(file_path)
    
    
##############@

def process_midi_files(base_folder):
    total_files = sum([len(files) for root, _, files in os.walk(base_folder) if any(file.lower().endswith('.mid') for file in files)])

    progress_bar = tqdm(total=total_files, desc="Processing MIDI files")
    
    for root, _, files in os.walk(base_folder):
        for file in files:
            if file.lower().endswith('.mid'):
                file_path = os.path.join(root, file)
                artist, song_title = extract_artist_song(file_path, base_folder)
                genre = find_genre_spotify(artist, song_title)

                if genre:
                    progress_bar.set_description(f"Adding genre '{genre}' to {file_path}")
                    add_genre_to_midi1(file_path, genre)
                else:
                    progress_bar.set_description(f"Could not find genre for {file_path}")

                progress_bar.update(1)
                time.sleep(1.1)  # Pause for 1.1 seconds between requests

    progress_bar.close()
    
    
##################

base_folder = '/Users/sho/Documents/VS/BETAauto_gpt_workspace/clean_midi'
process_midi_files(base_folder)
```

```python
import os
from mido import MidiFile
from tqdm import tqdm

def has_genre_data(file_path):
    try:
        midi_file = MidiFile(file_path)
        for track in midi_file.tracks:
            for message in track:
                if message.type == 'text' and message.text.startswith('Genre: '):
                    return True
    except Exception as e:
        print(f"Error processing file {file_path}: {str(e)}")
    return False

def collect_files_with_genre(base_folder, max_consecutive_non_genre_files=50):
    genre_files = []
    
    total_files = sum([len(files) for root, _, files in os.walk(base_folder) if any(file.lower().endswith('.mid') for file in files)])

    progress_bar = tqdm(total=total_files, desc="Checking genre information")

    non_genre_files_count = 0
    for root, _, files in os.walk(base_folder):
        for file in files:
            if file.lower().endswith('.mid'):
                file_path = os.path.join(root, file)
                if has_genre_data(file_path):
                    genre_files.append(file_path)
                    progress_bar.set_description(f"Found genre for {file_path}")
                    non_genre_files_count = 0
                else:
                    progress_bar.set_description(f"Genre not found for {file_path}")
                    non_genre_files_count += 1

                if non_genre_files_count >= max_consecutive_non_genre_files:
                    print(f"Stopped after {max_consecutive_non_genre_files} consecutive files without genre information.")
                    break

                progress_bar.update(1)
    
    progress_bar.close()
    return genre_files

base_folder = '/Users/sho/Documents/VS/BETAauto_gpt_workspace/clean_midi'
midi_files_with_genre = collect_files_with_genre(base_folder)
```
